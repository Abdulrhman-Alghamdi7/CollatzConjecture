```python
def f(x):
    n = 0
    m = 0
    while x > 1:
        if x % 2 == 0:
            x /= 2
        else:
            x = (x*3) + 1
        if x > m :
            m = x
        n+=1
            
    return [m,n]
```

```python
l = list()
n = int(input('enter the end of range : '))
for i in range(n):
    l.append((i,f(i)[0],f(i)[1]))
    z = 0
for i in l:
    z+=i[1]
z = z/len(l)
```

```python
if l[-1][0] != n-1:
    print('need to check : ' + l[-1][0]+1)
print(z)
```

```python
x = list()
y = list()
for i in l:
    x.append(i[0])
    y.append(i[1])

```

```python
import matplotlib.pyplot as plt
#Collatz Conjecture
plt.plot(x, y)
plt.axhline(y=z, color='red', linestyle='--', label='the mean of biggest values')
plt.xlabel('N values')
plt.ylabel('biggest values of N')
plt.title('Collatz Conjecture')
plt.show()
```
